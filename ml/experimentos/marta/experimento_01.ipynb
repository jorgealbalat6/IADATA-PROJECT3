```python
# Hoy many airbnbs do we have in each snapshot of listings
query = """
SELECT snapshot_date, COUNT(DISTINCT id) as num_pisos
FROM `project3grupo1.airbnb_raw.listings`
GROUP BY snapshot_date
ORDER BY snapshot_date
"""
print(client.query(query).to_dataframe())
```

```python
# How many rows are null in the calendar table and the proportion of free/occupied
query = """
SELECT 
  COUNT(*) as total,
  COUNTIF(price IS NULL) as price_nulo,
  COUNTIF(available = True) as ocupadas,
  COUNTIF(available = False) as libres
FROM `project3grupo1.airbnb_raw.calendar`
"""
print(client.query(query).to_dataframe())
```

```python
# How many dates do we have for each listing in calendar
query = """
SELECT 
  MIN(date) as fecha_min,
  MAX(date) as fecha_max,
  COUNT(DISTINCT listing_id) as num_pisos,
  COUNT(DISTINCT snapshot_date) as num_snapshots
FROM `project3grupo1.airbnb_raw.calendar`
"""
print(client.query(query).to_dataframe())
```

```python
query = """
SELECT applies_to_valencia, COUNT(*) as total
FROM `project3grupo1.airbnb_features.holidays`
GROUP BY applies_to_valencia
"""
print(client.query(query).to_dataframe())
```

```python
query = """
WITH listings_dedup AS (
    SELECT *
    FROM `project3grupo1.airbnb_raw.listings`
    QUALIFY ROW_NUMBER() OVER (PARTITION BY id ORDER BY snapshot_date DESC) = 1
),

events_agg AS (
    SELECT 
        start_date,
        COUNT(*) as num_events,
        MAX(phq_attendance) as max_attendance
    FROM `project3grupo1.airbnb_features.events`
    GROUP BY start_date
)

SELECT
    -- Target
    CASE WHEN c.available = False THEN 1 ELSE 0 END as ocupado,

    -- Temporales
    c.date,
    EXTRACT(DAYOFWEEK FROM c.date) as day_of_week,
    EXTRACT(MONTH FROM c.date) as month,
    EXTRACT(YEAR FROM c.date) as year,
    IF(EXTRACT(DAYOFWEEK FROM c.date) IN (1,7), 1, 0) as is_weekend,

    -- Del piso
    l.id as listing_id,
    l.neighbourhood_cleansed,
    l.room_type,
    l.accommodates,
    l.bedrooms,
    l.beds,
    l.price as listing_price,
    l.minimum_nights,
    l.number_of_reviews,
    l.review_scores_rating,
    l.review_scores_cleanliness,
    l.review_scores_location,
    l.review_scores_value,
    l.instant_bookable,
    l.latitude,
    l.longitude,

    -- Meteorología
    w.temp_max,
    w.temp_min,
    w.temp_mean,
    w.precipitation_mm,

    -- Eventos
    COALESCE(e.num_events, 0) as num_events,
    COALESCE(e.max_attendance, 0) as max_attendance,

    -- Festivos
    IF(h.date IS NOT NULL, 1, 0) as is_holiday

FROM `project3grupo1.airbnb_raw.calendar` c
LEFT JOIN listings_dedup l ON c.listing_id = l.id
LEFT JOIN `project3grupo1.airbnb_features.weather` w ON c.date = w.date
LEFT JOIN events_agg e ON c.date = e.start_date
LEFT JOIN `project3grupo1.airbnb_features.holidays` h 
    ON c.date = h.date AND h.applies_to_valencia = True
WHERE c.date BETWEEN '2025-06-23' AND '2026-04-30'
"""

df = client.query(query).to_dataframe()
print(df.shape)
print(df.isnull().sum())
```
